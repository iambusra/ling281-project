```python
from datasets import load_from_disk
ds = load_from_disk("hf_dataset")
```

```python
ds[0]
```

```python
# Print the top 10 authors who have written the most entries using ds['author']
# Counter 
from collections import Counter

# Count occurrences of each author
author_counts = Counter(ds['author'])

# Get the top 10 authors
top_authors = author_counts.most_common(10)

# Print the results
print("Top 10 authors by entry count:")
for author, count in top_authors:
    print(f"{author}: {count} entries")

# Print random entries from these top authors
import random
for author, count in top_authors:
    author_ds = ds.filter(lambda x: x['author'] == author)
    n_sample = min(3, len(author_ds))  # up to 3 entries per author
    if n_sample > 0:
        indices = random.sample(range(len(author_ds)), n_sample)
        sampled = author_ds.select(indices)
        print(f"\n--- {author} (sample of {n_sample}) ---")
        for row in sampled:
            snippet = row['content'][:150] + ("..." if len(row['content']) > 150 else "")
            topics_str = ", ".join(row.get('topics', []))
            suffix = f" ({topics_str})" if topics_str else ""
            print(f"  [{row['date_raw']}] {snippet}{suffix}")

```
